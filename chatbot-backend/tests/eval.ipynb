```python
from llama_index.core.evaluation import FaithfulnessEvaluator
from llama_index.core.evaluation import RelevancyEvaluator
from llama_index.core.evaluation import RetrieverEvaluator
from llama_index.llms.openai import OpenAI
# create llm
llm = OpenAI(model="gpt-4o-mini", temperature=0.0)

evaluator_faithfulness = FaithfulnessEvaluator(llm=llm)
evaluator_relevancy = RelevancyEvaluator(llm=llm)

# Danh sách để lưu các chỉ số
failfulness_scores = []
relevancy_scores = []

for i in range(len(question)):
    eval_result = evaluator_faithfulness.evaluate(query=str(question[i]),  contexts=context[i] , response=str(answer[i]))
    failfulness_score = eval_result.passing
    eval_result = evaluator_relevancy.evaluate(query=str(question[i]),  contexts=context[i] , response=str(answer[i]))
    relevancy_score = eval_result.passing
    # Thêm kết quả vào danh sách
    failfulness_scores.append(failfulness_score)
    relevancy_scores.append(relevancy_score)

# Tạo DataFrame với câu hỏi và các chỉ số đánh giá
pf = pd.DataFrame({
    'question': question,
    'faithfulness_score': failfulness_scores,
    'relevancy_score': relevancy_scores
})

# Hiển thị DataFrame
print(pf)
```
