```python
from llama_index.core.schema import Document
from llama_parse import LlamaParse

def _load_data(file_path: str) -> list[Document]:
    parser = LlamaParse(result_type="text")
    json_objs = parser.get_json_result(file_path)
    
    # In thông tin cơ bản của json_objs
    print("### json_objs Structure ###")
    print(f"Type: {type(json_objs)}")
    if isinstance(json_objs, list):
        print(f"Length: {len(json_objs)}")
        print(f"Keys of first item: {list(json_objs[0].keys())}")

    print("----------------------")
    
    json_list = json_objs[0]["pages"]  # Lấy danh sách các trang
    print("### json_list Structure ###")
    print(f"Type: {type(json_list)}")
    if isinstance(json_list, list):
        print(f"Length: {len(json_list)}")
        print(f"Keys of first item: {list(json_list[0].keys())}")
        print(f"Sample Data (First Item): {json_list[0]}")

    print("----------------------")
    
    docs = []
    for i, item in enumerate(json_list[9:20]):  # Lấy 10 trang đầu
        print(f"Page {i + 1} Keys: {list(item.keys())}")
        print(f"Metadata Type: {type(item['text'])} - Page Label Type: {type(item['page'])}")
        doc = Document(
            text=item["text"],
            metadata={"page_label": item["page"]},
        )
        docs.append(doc)

    print("### docs Structure ###")
    print(f"Type: {type(docs)}")
    print(f"Length: {len(docs)}")
    if docs:
        print(f"Sample Document Metadata: {docs[0].metadata}")
        print(f"Sample Document Text Type: {type(docs[0].text)}")
    
    return docs
```

```python
from sentence_transformers import SentenceTransformer, util
from llama_index.core.schema import Document
import re

# Tải mô hình embedding
model = SentenceTransformer('bkai-foundation-models/vietnamese-bi-encoder')  # Hoặc mô hình khác cho tiếng Việt nếu cần

def semantic_chunking(documents, threshold=0.75):
    chunks = []
    current_chunk = []
    current_embedding = None

    for document in documents:    
        sentences = document.text.split(r'[.?\n]')
        for sentence in sentences:
            sentence_embedding = model.encode(sentence, convert_to_tensor=True)

            # Nếu current_chunk rỗng, bắt đầu một chunk mới
            if current_chunk == []:
                current_chunk.append(sentence)
                current_embedding = sentence_embedding
            else:
                # Tính similarity với câu hiện tại
                similarity = util.cos_sim(current_embedding, sentence_embedding).item()
                if similarity > threshold:
                    current_chunk.append(sentence)  # Gộp vào chunk hiện tại
                    # Cập nhật embedding cho chunk hiện tại (trung bình các embedding)
                    current_embedding = (current_embedding + sentence_embedding) / 2
                else:
                    # Chunk đã kết thúc, lưu lại
                    chunks.append(" ".join(current_chunk))
                    # Bắt đầu chunk mới
                    current_chunk = [sentence]
                    current_embedding = sentence_embedding

    # Lưu chunk cuối cùng
    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

# Chunk tài liệu
chunks = semantic_chunking(documents, threshold=0.55)

# In ra các chunk
for i, chunk in enumerate(chunks):
    print(f"Chunk {i + 1}:")
    print(chunk)
    print("-" * 50)
```
